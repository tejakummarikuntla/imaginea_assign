### Gaussian Probability Density Function.

$$f(x; \mu, \sigma) = \frac{1}{\sqrt{2 \sigma^2 \pi}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$

**Using Scipy**
```python
from scipy.stats import norm
norm.pdf(x, mu, sigma)
```

````python
plt.plot(sorted(r),norm.pdf(sorted(r)))
````

So if we want to see the probability of 2 and 6 are drawn from a distribution with ** = 4** and ** = 1** we get:

````python
norm.pdf(2, 4, 1)*norm.pdf(6, 4, 1)
````
