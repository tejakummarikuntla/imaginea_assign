```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("ner_dataset.csv", encoding="latin1")
```

```python
data.head()
```

```python
# Calculating Initial Probabilites

inital_prob_df = data.Tag.value_counts()
inital_prob_df = pd.DataFrame(inital_prob_df)
inital_prob_df['initial_porb'] = inital_prob_df.Tag/inital_prob_df.Tag.sum()
inital_prob_df
```

```python
# Creating Transition Matrix
transition_mat = pd.DataFrame(columns=inital_prob_df.Tag.keys(), index=inital_prob_df.Tag.keys())
transition_mat
```

```python
# Adding i+1 Tag value as a new Column

temp_df = data.Tag.iloc[1:]
temp_df_cast_list = list(temp_df)
temp_df_cast_list.append(np.nan)
```

```python
len(temp_df_cast_list)
```

```python
len(data)
```

```python
data['Next_Tag'] = temp_df_cast_list
data.head(10)
```

```python
data.head(100)
```

```python
sent_index = data[data['Sentence #'].notnull()].index
sent_index
```

```python
# Substracing -1 from the recived indexes.
def substract_non_zero(lst):
    res = []
    for item in lst:
        if item == 0:
            res.append(item)
            continue
        else:
            res.append(item-1)
    return res

sent_index_above = []
sent_index_above = substract_non_zero(sent_index)
```

```python
sent_index_above[:10]
```

```python
# adding column Index

data['Index'] = data.index
```

```python
def rm_tag(row):
    if row['Index'] in sent_index_above:
        row['Next_Tag'] = np.nan
    return row
```

```python
data.apply(rm_tag, axis=1)
```
